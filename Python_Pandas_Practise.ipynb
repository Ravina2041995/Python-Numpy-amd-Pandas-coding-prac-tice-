# Making changes in Data

```python
poke.head(5)
```

```python
#Adding different column and make into one 
poke['Total'] = poke['HP']+poke['Attack']+poke['Defense']+poke['Sp. Atk']+poke['Sp. Def']+poke['Speed']
poke.head(2)
```

```python
#Droping some column, always save value to a framework, so that if doesn't modift the original one
poke=poke.drop(columns=['Total'])
poke.head(2)
```

```python
#Axis =0, will add all values vertically, 
#iloc: first value is for row and second start and end of column
poke['Total'] = poke.iloc[:,4:9].sum(axis =1)
poke.head(2)
```

```python
#Shifting total to different column
cols = list(poke.columns)
poke  = poke[cols[0:4]+[cols[-1]]+cols[4:12]]
#0:4- starting  4 columns should be same
#-1: last column we want to add after 4
#4:12- Summation of 4 to 12
poke
```

# Filtering Data

```python
#Type 1 with only grass
poke.loc[poke['Type 1'] == 'Grass']
```

```python
# With 2 conditions 
poke.loc[(poke['Type 1'] == 'Grass') & (poke['Type 2'] == 'Poison')]
```

```python
# With or conditions 
new_poke= poke.loc[(poke['Type 1'] == 'Grass') | (poke['Type 2'] == 'Poison')]

#Reset your index
# new_poke = new_poke.reset_index()

#Reset your index with no new column
new_poke = new_poke.reset_index(drop = True)
new_poke

#Reset your index with no new column and don't want to create a new df
new_poke.reset_index(drop = True, inplace = True)
new_poke
```

```python
#Filter name with all mega version in it
poke.loc[poke['Name'].str.contains('Mega')]
```

```python
#Filter name with not any mega version in it
poke.loc[~poke['Name'].str.contains('Mega')]
```

# Complicated filtering with regression function

```python
import re
#Import data with column Type 1 which contains only fire or grass
#To ignore capital letter include flags=re.I 
poke.loc[poke['Type 1'].str.contains('fire|grass',flags=re.I, regex=True)]
```

```python
#Filter name with all starting with 'pi' and 
poke.loc[poke['Name'].str.contains('^pi[a-z]*',flags=re.I, regex=True)]
```

# Conditional Changes

```python
#Change fire in type 1 column to flamer
poke.loc[poke['Type 1']== 'Fire','Type 1'] = 'Flamer'
poke
```

```python
#Change fire column with new column and give value to iy
poke.loc[poke['Type 1']== 'Fire','Legendary'] = True
poke
```

```python
#If condition meets, change 2 column value to specific value 
poke.loc[poke['Total']>500,['Generation','Legendary']]=['Test 1','Test 2']
poke
```

# Aggregate Statistic 
(GroupBy)

```python
poke.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
```

```python
poke.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)
poke.groupby(['Type 1']).sum()
poke.groupby(['Type 1']).count()
```

```python
poke['count'] =1
poke.groupby(['Type 1']).count()['count']
```

```python
#Multiple parameter groupby
poke.groupby(['Type 1', 'Type 2']).count()['count']
```

# Working with Large amount of data

```python
for df in pd.read_csv('modified.csv',chunksize = 5):
    print(df)
    
```

```python
new_poke = pd.DataFrame(columns = poke.columns)
for df in pd.read_csv('modified.csv',chunksize =5):
    results = df.groupby(['Type 1']).count()
    new_poke = pd.concat([new_poke,results])
```
